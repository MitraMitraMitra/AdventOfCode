```python
class point():
    def __init__(self, risk):
        self.risk = risk
        self.risk_distance = None
        self.risk_path = []
```

```python
matrix_points = []
for i in matrix:
    row = []
    for j in i:
        row.append(point(j))
    matrix_points.append(row)
```
