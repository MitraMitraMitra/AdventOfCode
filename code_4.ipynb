```python
import pandas as pd
import re
```

```python
numbers = []
f = open('puzzle_4_input.txt', 'r')
for i in f:
    numbers.append(i)
```

```python
draw_numbers = numbers[0].split(',')
draw_numbers[-1] = draw_numbers[-1][0:-1]
draw_numbers = [int(i) for i in draw_numbers]
numbers = numbers[1:]
print(draw_numbers)
```

```python
boards = []
board = []
for i in range(1, len(numbers)):
    if i % 6 != 0:
        row = numbers[i][0:-1]
        row = re.findall('[0-9]+', row)
        row = [int(i) for i in row]
        board.append(row)
    else:
        boards.append(pd.DataFrame(board))
        board = []
```

```python
def mark_board(df, x):
    for i in range(0, len(df)):
        for j in df.columns:
            try:
                if int(df[j][i]) == x:
                    df[j][i] = None
            except:
                pass
    return df
```

```python
def check_board(df):
    for i in range(0, len(df)):
        x = df[i].unique()
        if str(x) == '[nan]':
            return True
    for j in df.columns:
        x = df.loc[j].unique()
        if str(x) == '[nan]':
            return True
    return False
```

```python
won = []
for i in draw_numbers:
    for j in range(0, len(boards)):
        boards[j] = mark_board(boards[j], i)
        if check_board(boards[j]):
            if j not in won:
                print('board[', j, ']: ', boards[j].sum().sum() * i, sep = '')
                won.append(j)
            #boards.remove(boards[j])
        
```
