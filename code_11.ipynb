```python
def flash(x, y):
    global matrix
    global total_flashes
    total_flashes = total_flashes + 1
    #matrix[x][y] = matrix[x][y] - 10
    matrix[x][y] = 0
    #print('touched by the flash of (', x, ', ', y, '):', sep = '')
    for i in range(0, len(matrix)):
        for j in range(0, len(matrix)):
            if (i != x or j != y) and math.sqrt((x - i) ** 2 + (y - j) ** 2) < 2:
                #print('matrix[', i, '][', j, '] (', matrix[i][j], ') will now be increased by 1', sep = '')
                matrix[i][j] = matrix[i][j] + 1
                if matrix[i][j] > 9:
                    flash(i, j)
```

```python
def flash2(x, y):
    global matrix
    global total_flashes
    total_flashes = total_flashes + 1
    #matrix[x][y] = matrix[x][y] - 10
    matrix[x][y] = 0
    #print('touched by the flash of (', x, ', ', y, '):', sep = '')
    to_flash = []
    for i in range(0, len(matrix)):
        for j in range(0, len(matrix)):
            if (i != x or j != y) and math.sqrt((x - i) ** 2 + (y - j) ** 2) < 2:
                #print('matrix[', i, '][', j, '] (', matrix[i][j], ') will now be increased by 1', sep = '')
                if (i, j) not in to_flash:
                    matrix[i][j] = matrix[i][j] + 1
                    if matrix[i][j] > 9:
                        to_flash.append((i, j))
    for i in to_flash:
        flash(i[0], i[1])
```
